```python
from gdeltdoc import GdeltDoc, Filters
import pandas as pd

# Функция для получения timeline по ключевым словам
def get_timeline(keywords, start_date="2017-01-01", end_date="2025-12-01"):
    f = Filters(
        start_date=start_date,
        end_date=end_date,
        keyword=keywords
    )
    gd = GdeltDoc()
    timeline = gd.timeline_search("timelinetone", f)
    # Преобразуем в DataFrame
    df = pd.DataFrame(timeline, columns=["datetime", "Average Tone"])
    #df["keyword"] = ", ".join(keywords)
    return df

# Списки ключевых слов для каждого объекта
keywords_list = {
    "Putin": ["Putin","Путин","кремль","kremlin"],
    "Trump": ["Trump","Трамп"],
    "Biden": ["Biden","Байден"],
    "Charlie Kirk": ["Charlie Kirk","Чарли Кирк"],
    "Li Qiang": ["Li Qiang","Ля Цяэ"],
    "Xi Jinping": ["Xi Jinping","Си Цзиньпин"]
}

# Собираем данные по всем
dfs = []
for name, kws in keywords_list.items():
    df = get_timeline(kws)
    df["entity"] = name  # для удобства
    dfs.append(df)

# Объединяем в один DataFrame
final_df = pd.concat(dfs, ignore_index=True)

# Проверка
#final_df.head()
# Экспорт в CSV
final_df.to_csv("gdelt_timelinetone.csv", index=False, encoding="utf-8")
```
